```python
#Import libraries 
import xarray as xr
import matplotlib.pyplot as plt
```

```python
# Load INSAT-3DR data file
data = xr.open_mfdataset('H:/Jul24_104145/AUG2020/3RSND_*.h5',concat_dim='time',combine='nested', engine="h5netcdf")
data
```

```python
# Multiplying factor to Latitude and Longitude
data.assign_coords(Longitude=(data.Longitude *0.01))
data.assign_coords(Latitude=(data.Latitude *0.01))
data
```

```python
# clipping the data between the area
min_lon = 72
min_lat = 17 
max_lon = 74
max_lat = 19

da1 = data.TAirPhy.where((data.Latitude > min_lat).compute() & (data.Latitude < max_lat).compute(), drop=True)
da2 = da1.where((da1.Longitude > min_lon).compute() & (da1.Longitude < max_lon).compute(), drop=True)
da2
```

```python
# Spatial mean
data_mean = da2.mean(dim=['GeoX','GeoY'], skipna=True)
data_mean
```
